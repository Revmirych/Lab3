```python
def hanoi(n, from_peg, to_peg, spare_peg):
  if n == 1:
    print("Переместить диск ", n," с", from_peg, "на", to_peg)
  else:
    hanoi(n - 1, from_peg, spare_peg, to_peg)
    print("Переместить диск ", n," с", from_peg, "на", to_peg)
    hanoi(n - 1, spare_peg, to_peg, from_peg)


n = 9
from_peg = "A"
to_peg = "C"
spare_peg = "B"

hanoi(n, from_peg, to_peg, spare_peg)
```
